

## 🧠 NumPy (Numerical Python) — Sayısal Hesaplama Motoru

### 📌 1. NumPy Nedir?

NumPy, Python’da sayısal verilerle çalışmak için geliştirilmiş bir kütüphanedir. Temel veri yapısı **ndarray** adı verilen çok boyutlu dizilerdir.

### 🔧 Neden Kullanılır?

* **Çok boyutlu dizi (array)** oluşturma ve üzerinde işlem yapma
* **Matematiksel işlemler** (toplama, çarpma, logaritma, trigonometrik işlemler vs.)
* **Lineer cebir** (matris çarpımı, determinant, ters alma vs.)
* **İstatistiksel hesaplamalar** (ortalama, medyan, varyans)
* **Performans avantajı**: Python listelerine göre çok daha hızlıdır çünkü C diliyle yazılmıştır.

---

### 📎 NumPy Temel Örnekler:




````python
import numpy as np

# Tek boyutlu dizi
a = np.array([1, 2, 3, 4])
print(a * 2)  # [2 4 6 8]


````

````python
# 2D matris
b = np.array([[1, 2], [3, 4]])
b

````

````python
print(np.transpose(b))  # [[1 3] [2 4]]
````

````python
# Rastgele sayı üretimi
np.random.seed(0)
print(np.random.randint(0, 10, size=(2, 3)))  # 2x3 boyutlu rastgele tam sayılar
````

````python

# İstatistiksel işlemler
print(np.mean(a))  # Ortalama
````

````python
print(np.std(a))   # Standart sapma
````

## 📊 Pandas — Veri Analizi ve Temizleme Aracı

### 📌 2. Pandas Nedir?

Pandas, yapılandırılmış verilerle (CSV, Excel, SQL, JSON vb.) çalışmayı kolaylaştıran bir kütüphanedir. Özellikle veri analizi, temizleme, dönüştürme ve özetleme gibi işlemler için çok kullanılır.

### 🔧 Neden Kullanılır?

* **Tablolu verilerle çalışma (DataFrame)** — tıpkı Excel gibi
* **Veri temizleme** — eksik veri silme, doldurma
* **Filtreleme ve sıralama**
* **Gruplama ve özetleme** (groupby)
* **Zaman serisi analizleri**
* **CSV/Excel/SQL okuma-yazma**

---

### 📎 Pandas Temel Örnekler:

```python
import pandas as pd

# Basit bir DataFrame oluşturma
data = {
    "İsim": ["Ali", "Ayşe", "Mehmet"],
    "Yaş": [25, 30, 22],
    "Şehir": ["İstanbul", "Ankara", "İzmir"]
}
df = pd.DataFrame(data)

# Filtreleme: Yaşı 24'ten büyük olanlar
print(df[df["Yaş"] > 24])

# Yeni sütun ekleme
df["Yeni Puan"] = [85, 90, 78]

# Ortalama yaş
print(df["Yaş"].mean())

# CSV'den veri okuma
# df = pd.read_csv("veriler.csv")

# Gruplama örneği
grouped = df.groupby("Şehir")["Yaş"].mean()
print(grouped)
```

---

## 🔍 NumPy vs Pandas

| Özellik      | NumPy                       | Pandas                         |
| ------------ | --------------------------- | ------------------------------ |
| Veri Yapısı  | ndarray (dizi)              | Series, DataFrame (etiketli)   |
| Amaç         | Sayısal işlemler, matematik | Veri analizi, tablo işlemleri  |
| Hız          | Daha hızlı                  | Daha kullanıcı dostu           |
| Veri Kaynağı | Sayısal veri (genelde)      | Her türlü yapılandırılmış veri |

---

## ✅ Özet

````python
x = np.array(df["TV"])
y = np.array(df["sales"])

# En küçük kareler yöntemiyle basit doğrusal regresyon
a, b = np.polyfit(x, y, 1)  # a: eğim, b: kesişim noktası

plt.scatter(x, y, label="Veri")
plt.plot(x, a * x + b, color='red', label=f"y = {a:.2f}x + {b:.2f}")
plt.xlabel("TV Reklam Harcaması")
plt.ylabel("Satışlar")
plt.title("TV Reklamı ile Satış İlişkisi")
plt.legend()
plt.show()
````
