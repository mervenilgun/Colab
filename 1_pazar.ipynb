

## üß† NumPy (Numerical Python) ‚Äî Sayƒ±sal Hesaplama Motoru

### üìå 1. NumPy Nedir?

NumPy, Python‚Äôda sayƒ±sal verilerle √ßalƒ±≈ümak i√ßin geli≈ütirilmi≈ü bir k√ºt√ºphanedir. Temel veri yapƒ±sƒ± **ndarray** adƒ± verilen √ßok boyutlu dizilerdir.

### üîß Neden Kullanƒ±lƒ±r?

* **√áok boyutlu dizi (array)** olu≈üturma ve √ºzerinde i≈ülem yapma
* **Matematiksel i≈ülemler** (toplama, √ßarpma, logaritma, trigonometrik i≈ülemler vs.)
* **Lineer cebir** (matris √ßarpƒ±mƒ±, determinant, ters alma vs.)
* **ƒ∞statistiksel hesaplamalar** (ortalama, medyan, varyans)
* **Performans avantajƒ±**: Python listelerine g√∂re √ßok daha hƒ±zlƒ±dƒ±r √ß√ºnk√º C diliyle yazƒ±lmƒ±≈ütƒ±r.

---

### üìé NumPy Temel √ñrnekler:




````python
import numpy as np

# Tek boyutlu dizi
a = np.array([1, 2, 3, 4])
print(a * 2)  # [2 4 6 8]


````

````python
# 2D matris
b = np.array([[1, 2], [3, 4]])
b

````

````python
print(np.transpose(b))  # [[1 3] [2 4]]
````

````python
# Rastgele sayƒ± √ºretimi
np.random.seed(0)
print(np.random.randint(0, 10, size=(2, 3)))  # 2x3 boyutlu rastgele tam sayƒ±lar
````

````python

# ƒ∞statistiksel i≈ülemler
print(np.mean(a))  # Ortalama
````

````python
print(np.std(a))   # Standart sapma
````

## üìä Pandas ‚Äî Veri Analizi ve Temizleme Aracƒ±

### üìå 2. Pandas Nedir?

Pandas, yapƒ±landƒ±rƒ±lmƒ±≈ü verilerle (CSV, Excel, SQL, JSON vb.) √ßalƒ±≈ümayƒ± kolayla≈ütƒ±ran bir k√ºt√ºphanedir. √ñzellikle veri analizi, temizleme, d√∂n√º≈üt√ºrme ve √∂zetleme gibi i≈ülemler i√ßin √ßok kullanƒ±lƒ±r.

### üîß Neden Kullanƒ±lƒ±r?

* **Tablolu verilerle √ßalƒ±≈üma (DataFrame)** ‚Äî tƒ±pkƒ± Excel gibi
* **Veri temizleme** ‚Äî eksik veri silme, doldurma
* **Filtreleme ve sƒ±ralama**
* **Gruplama ve √∂zetleme** (groupby)
* **Zaman serisi analizleri**
* **CSV/Excel/SQL okuma-yazma**

---

### üìé Pandas Temel √ñrnekler:

```python
import pandas as pd

# Basit bir DataFrame olu≈üturma
data = {
    "ƒ∞sim": ["Ali", "Ay≈üe", "Mehmet"],
    "Ya≈ü": [25, 30, 22],
    "≈ûehir": ["ƒ∞stanbul", "Ankara", "ƒ∞zmir"]
}
df = pd.DataFrame(data)

# Filtreleme: Ya≈üƒ± 24'ten b√ºy√ºk olanlar
print(df[df["Ya≈ü"] > 24])

# Yeni s√ºtun ekleme
df["Yeni Puan"] = [85, 90, 78]

# Ortalama ya≈ü
print(df["Ya≈ü"].mean())

# CSV'den veri okuma
# df = pd.read_csv("veriler.csv")

# Gruplama √∂rneƒüi
grouped = df.groupby("≈ûehir")["Ya≈ü"].mean()
print(grouped)
```

---

## üîç NumPy vs Pandas

| √ñzellik      | NumPy                       | Pandas                         |
| ------------ | --------------------------- | ------------------------------ |
| Veri Yapƒ±sƒ±  | ndarray (dizi)              | Series, DataFrame (etiketli)   |
| Ama√ß         | Sayƒ±sal i≈ülemler, matematik | Veri analizi, tablo i≈ülemleri  |
| Hƒ±z          | Daha hƒ±zlƒ±                  | Daha kullanƒ±cƒ± dostu           |
| Veri Kaynaƒüƒ± | Sayƒ±sal veri (genelde)      | Her t√ºrl√º yapƒ±landƒ±rƒ±lmƒ±≈ü veri |

---

## ‚úÖ √ñzet
