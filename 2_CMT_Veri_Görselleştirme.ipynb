## **Öğrenci Notları Veri Seti:**





````python

import pandas as pd

# Veri seti
data = {
    "Ad": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet", "Fatma", "Deniz", "Ece"],
    "Not": [65, 90, 78, 85, 92, 70, 58, 88]
}

df = pd.DataFrame(data)
````

````python
df
````


### **1. Histogram (Dağılımı Gösterir)**






````python
import matplotlib.pyplot as plt

plt.hist(df["Not"], bins=5, color='skyblue', edgecolor='black')
plt.title("Öğrenci Notlarının Dağılımı")
plt.xlabel("Not Aralığı")
plt.ylabel("Öğrenci Sayısı")
plt.grid(axis='y', linestyle='--', color='gray', alpha=0.6)
plt.show()
````

Yorum:
Histogram, notların hangi aralıklarda yoğunlaştığını gösterir.
Notlar hangi aralıkta yığılmış durumda? Not dağılımı normal mi?

### **2. Bar Chart (Kişiye Göre Not Karşılaştırması)**




**Yorum:**
Bar chart, bireysel karşılaştırma yapmamıza olanak sağlar.

> En yüksek ve en düşük not hangi öğrenciye ait?

````python

plt.bar(df["Ad"], df["Not"], color='orange')
plt.title("Öğrencilerin Notları")
plt.xlabel("Öğrenciler")
plt.ylabel("Not")
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()
````

### **3. Box Plot (İstatistiksel Özet ve Aykırı Değerler)**



**Yorum:**

> Notlar ortalamaya yakın mı? Dağılım simetrik mi?

**Yorum:**

````python
import matplotlib.pyplot as plt

# Veriler
aylar = ['Ocak', 'Şubat', 'Mart', 'Nisan', 'Mayıs', 'Haziran']
gelir = [3200, 3400, 3100, 3600, 3900, 4100]
gider = [2800, 2900, 3000, 3100, 3300, 3500]

# Kâr = Gelir - Gider
kar = [g - h for g, h in zip(gelir, gider)]
harcama = gider  # Aynı veri, isimlendirme farkı

# Grafikler
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# 1. grafik: Kâr çizgi grafiği
axes[0].plot(aylar, kar, color='purple', marker='o')
axes[0].set_title("Aylık Kâr")
axes[0].set_ylabel("Tutar (TL)")
axes[0].grid(True)

# 2. grafik: Harcama bar chart
axes[1].bar(aylar, harcama, color='orange')
axes[1].set_title("Aylık Harcama")
axes[1].set_ylabel("Tutar (TL)")
axes[1].grid(axis='y')

plt.tight_layout()
plt.show()
````

## 2. Barplot (Ortalama Değer Gösterimi)
**Yorum:**
Seaborn barplot, ortalamaları gösterir. ci=None ile güven aralığı kapatılır.

````python
import warnings # warningleri kapatma
warnings.filterwarnings("ignore")
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips, ci=None, palette="Blues")#ci güven aralığı veren parametre. akademik çalışmalarda 95 değeri ile kullanılır
plt.title("Günlere Göre Ortalama Hesap")
plt.show()
````

## 3. Boxplot (Dağılım ve Aykırı Değer)

````python
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("Günlere Göre Hesap Dağılımı")
plt.show()
````

## 4. Histogram / KDE Plot

**Yorum:**
Histogram, verinin dağılımını gösterir. kde=True ile eğri çizilir. Yoğunluk analizinde kullanılır.

````python
sns.histplot(data=tips, x="total_bill", kde=True, bins=20, color="orange")
plt.title("Toplam Hesap Dağılımı")
plt.show()
````

##5. Scatterplot (İki Sürekli Değişken Arasındaki İlişki)

Scatterplot, iki sayısal değişken arasındaki ilişkiyi gösterir. hue parametresiyle renk ayrımı sağlanır.

````python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
plt.title("Toplam Hesap vs Bahşiş")
plt.show()
````

## 6. Heatmap (Korelasyon Gösterimi)
heatmap, değişkenler arasındaki korelasyonu görselleştirir. Özellikle feature selection öncesi kullanılır.



````python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Korelasyon Matrisi")
plt.show()
````

## 7. Seaborn Tema ve Palet Kullanımı

**Yorum:**
Görselleştirmenin etkileyiciliği tema ve renk seçimiyle artar.

````python
sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(x="day", y="total_bill", data=tips, ax=axes[0])
axes[0].set_title("Boxplot")

sns.histplot(data=tips, x="total_bill", kde=True, ax=axes[1])
axes[1].set_title("Histogram + KDE")

plt.tight_layout()
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Veri setini yükle
tips = sns.load_dataset("tips")

# Şekil ve alt grafikler
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1️⃣ Boxplot – Günlere göre hesap dağılımı
sns.boxplot(x="day", y="total_bill", data=tips, ax=axes[0], palette="pastel")
axes[0].set_title("Günlere Göre Ödenen Hesap Dağılımı")
axes[0].set_xlabel("Gün")
axes[0].set_ylabel("Toplam Hesap (TL)")

# 2️⃣ Bar chart – Günlere göre kişi sayısı
sns.countplot(x="day", data=tips, ax=axes[1], palette="Blues")
axes[1].set_title("Günlere Göre Kişi Sayısı")
axes[1].set_xlabel("Gün")
axes[1].set_ylabel("Kişi Sayısı")

plt.tight_layout()
plt.show()
````

## Interaktif Bar Chart + Hover Tool

````python
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure, show

aylar = ['Ocak', 'Şubat', 'Mart', 'Nisan']
satis = [120, 150, 130, 170]
source = ColumnDataSource(data=dict(ay=aylar, satis=satis))

p = figure(x_range=aylar, title="Aylık Satışlar", tools="")

p.vbar(x='ay', top='satis', width=0.5, source=source, color="orange")

hover = HoverTool(tooltips=[("Ay", "@ay"), ("Satış", "@satis")])
p.add_tools(hover)

show(p)
````

# Veri Biliminde Veri Görselleştirmenin Yeri



## **1. Veri Görselleştirme = İlk Temas, İlk İçgörü**

Makine öğrenmesine başlamadan önce **veriyi tanımak** zorundayız.
Görselleştirme, bu tanışmanın ilk adımıdır:

* Dağılım nedir?
* Eksik veriler var mı?
* Aykırı değerler (outlier) kim?
* Sınıflar dengeli mi?
* Değişkenler arasında ilişki var mı?

**Örnek:**
Sınıflandırma yapacaksın ama dağılım dengesiz (imbalanced classes) → `sns.countplot()`
Bu olmadan modeli eğitirsen model yüksek accuracy verip hatalı genelleme yapabilir.

---

## **2. Özellik (Feature) Seçimi ve Mühendisliği İçin Temel Girdidir**

Model başarısı büyük ölçüde **doğru ve anlamlı feature'lara** bağlıdır.

* Korelasyon analizi → `heatmap` ile: hangi değişkenler çok benzer?
* Scatter plot → lineer mi, nonlineer mi?
* Boxplot → uç değer problemi var mı?

**Örnek:**
Petal length ve petal width çok yüksek korelasyon gösteriyorsa → biri elenebilir veya PCA uygulanabilir.

---

## **3. Model Performansı ve Hataların Görselleştirilmesi**

Bir model eğittik, ama iyi mi kötü mü?

* Confusion Matrix → Hangi sınıflar karışıyor?
* ROC Curve → Sınıflandırıcının başarısı
* Loss eğrisi → Overfitting var mı?

**Örnek:**
Line chart ile epoch'lara göre `val_loss` görselleştirmek → erken durdurma (early stopping) için kritik karar noktasıdır.

---

## **4. Paydaşlarla İletişim İçin Evrensel Dil**

Bir yöneticiyi, yatırımcıyı, müşteriyi `f1-score` ile ikna edemezsin. Ama **grafiklerle anlatabilirsin.**

* Interaktif grafik → kullanıcı veriyi kendisi deneyimler
* Dashboard → hızlı karar alma süreçlerinde temel araç

**Plotly ve Bokeh** burada devreye girer. Hem **veriye hâkimiyeti gösterir**, hem **sunumu güçlendirir**.

---

## **5. Modelin Anlaşılabilirliği (Explainability)**

Model neye göre karar veriyor? Feature importance nedir? SHAP değerleri nedir?

Bunların hepsi grafikle görselleştirilir.

---

## **Özetle:**

| Amaç             | Kullanım                                |
| ---------------- | --------------------------------------- |
| Veriyi Tanıma    | Histogram, Scatter, Bar, Box            |
| Feature Analizi  | Heatmap, Correlation Matrix             |
| Model Hataları   | Confusion Matrix, ROC Curve, Loss Graph |
| İletişim         | Plotly, Bokeh ile Dashboard             |
| Model Açıklaması | SHAP, Partial Dependence Plot           |

---

**Son söz:**

> **Veri görselleştirme, verinin dilidir.**
> Onu konuşabilen veri bilimciler, modellerini sadece eğitmez — anlar, anlatır ve aktarır.

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
output_notebook()

df = sns.load_dataset("iris")
````

### 1. Matplotlib - Bar Chart: Türlere Göre Sayı Dağılımı



````python
df['species'].value_counts().plot(kind='bar', color=['skyblue', 'orange', 'lightgreen'])
plt.title("Çiçek Türlerine Göre Sayı")
plt.xlabel("Tür")
plt.ylabel("Adet")
plt.grid(True)
plt.show()
````

````python
plt.hist(df["sepal_length"], bins=10, color='orange', edgecolor='black')
plt.title("Sepal Length Dağılımı")
plt.xlabel("Sepal Length")
plt.ylabel("Frekans")
plt.grid(True)
plt.show()
````

````python
avg_petal = df.groupby("species")["petal_width"].mean().reset_index()
fig = px.bar(avg_petal, x="species", y="petal_width", color="species",
             title="Türlere Göre Ortalama Petal Width")
fig.show()
````

### 8. Bokeh - Interaktif Bar Chart: Sepal Width Ortalaması



````python
means = df.groupby("species")["sepal_width"].mean().reset_index()
source = ColumnDataSource(means)

p = figure(x_range=means["species"], title="Sepal Width Ortalamaları", tools="")
p.vbar(x="species", top="sepal_width", source=source, width=0.5, color="navy")

hover = HoverTool(tooltips=[("Tür", "@species"), ("Ortalama Genişlik", "@sepal_width")])
p.add_tools(hover)

show(p)
````
