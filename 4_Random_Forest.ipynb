

## üéØ Random Forest ve Hiperparametre Ayarlarƒ±

---

### üìå 1. Giri≈ü: Random Forest Nedir?

**Ana Fikir:**  
Random Forest, bir√ßok karar aƒüacƒ±nƒ±n birlikte √ßalƒ±≈ütƒ±ƒüƒ± bir topluluk (ensemble) √∂ƒürenme y√∂ntemidir.  
Her bir aƒüa√ß verinin farklƒ± bir alt k√ºmesinde ve √∂zelliklerin rastgele se√ßilen bir kƒ±smƒ±nda eƒüitilir.  
Sonu√ßta aƒüa√ßlar oy verir (sƒ±nƒ±flandƒ±rma i√ßin √ßoƒüunluk oyu, regresyon i√ßin ortalama alƒ±nƒ±r).

**√ñrnek benzetme:**  
Random Forest = Kalabalƒ±k bir uzmanlar grubu ‚Üí Her biri farklƒ± bir bakƒ±≈ü a√ßƒ±sƒ± ‚Üí Toplu karar ‚Üí Daha doƒüru ve dengeli sonu√ß.

---

### üß† 2. Random Forest'ƒ±n Faydalarƒ±

- A≈üƒ±rƒ± √∂ƒürenmeye (overfitting) kar≈üƒ± dayanƒ±klƒ±dƒ±r.
- Karar aƒüacƒ±ndan daha kararlƒ± ve genelleme yeteneƒüi y√ºksektir.
- Feature importance (√∂zellik √∂nem sƒ±ralamasƒ±) sunar.
- Hem sƒ±nƒ±flandƒ±rma hem regresyon problemlerinde kullanƒ±labilir.

---

### ‚öôÔ∏è 3. Hiperparametreler

| Parametre | A√ßƒ±klama | Not |
|-----------|----------|-----|
| `n_estimators` | Olu≈üturulacak aƒüa√ß sayƒ±sƒ± | Genelde ne kadar √ßok, o kadar iyi (ama yava≈ülar) |
| `max_depth` | Her aƒüacƒ±n derinliƒüi | A≈üƒ±rƒ± √∂ƒürenmeyi engellemek i√ßin sƒ±nƒ±rlanabilir |
| `min_samples_split` | Bir dalƒ±n b√∂l√ºnebilmesi i√ßin gereken minimum √∂rnek sayƒ±sƒ± | Daha y√ºksek ‚Üí daha basit model |
| `min_samples_leaf` | Bir yaprak d√ºƒü√ºmdeki minimum √∂rnek sayƒ±sƒ± | Daha y√ºksek ‚Üí daha genel aƒüa√ßlar |
| `max_features` | Her aƒüa√ß i√ßin deƒüerlendirilecek maksimum √∂zellik sayƒ±sƒ± | `sqrt`, `log2`, ya da sayƒ± olabilir |
| `bootstrap` | √ñrnekleme y√∂ntemi (`True` ‚Üí √∂rneklerle, `False` ‚Üí t√ºm veri) | |

## üå≤ Random Forest Nasƒ±l √áalƒ±≈üƒ±r?

### üîÅ 1. Temel Fikir: **Topluluk √ñƒürenmesi (Ensemble Learning)**

Random Forest, "bir karar iyi olabilir ama bir√ßok farklƒ± kararƒ± birle≈ütirmek genelde daha iyidir" ilkesine dayanƒ±r.  
Bu y√ºzden *bir√ßok karar aƒüacƒ±* olu≈üturur ve onlarƒ±n tahminlerini birle≈ütirir.

- **Sƒ±nƒ±flandƒ±rma** i√ßin: √áoƒüunluk oyu (majority voting)
- **Regresyon** i√ßin: Tahminlerin ortalamasƒ± (average)

---

### üîç 2. Random Forest Adƒ±mlarƒ±

#### üß© Adƒ±m 1: **Bootstrap √ñrnekleme (Veri Alt K√ºmesi Se√ßimi)**

- Eƒüitim verisinden *tesad√ºfi olarak* (geri koymalƒ± ≈üekilde) bir alt k√ºme se√ßilir.
- Her aƒüa√ß, kendi √∂rnekleme verisiyle eƒüitilir.  
  ‚Üí Bu y√ºzden her aƒüa√ß farklƒ± bir "bakƒ±≈ü a√ßƒ±sƒ±" geli≈ütirir.

> √ñrnek: 100 g√∂zlemden olu≈üan bir veri seti i√ßin her aƒüa√ß 100 rastgele (ve tekrar edebilir) √∂rnek alƒ±r.

---

#### üé≤ Adƒ±m 2: **Rasgele √ñzellik Se√ßimi (Feature Bagging)**

- Her d√ºƒü√ºm (node) i√ßin:
  - T√ºm √∂zellikler arasƒ±ndan *rastgele se√ßilen k√º√ß√ºk bir alt k√ºme* deƒüerlendirilir.
  - Bu √∂zellik k√ºmesinden en iyi b√∂l√ºnme se√ßilir.

> Bu sayede aƒüa√ßlar *birbirine benzememeye zorlanƒ±r*, √ße≈üitlilik artar.

---

#### üå≥ Adƒ±m 3: **Karar Aƒüa√ßlarƒ±nƒ±n Eƒüitilmesi**

- Bootstrap verisi ve rastgele √∂zelliklerle bir√ßok karar aƒüacƒ± (√∂rneƒüin 100 aƒüa√ß) eƒüitilir.
- Her aƒüa√ß ‚Äútam bir model‚Äùdir ama diƒüerlerinden farklƒ± √∂ƒürenmi≈ütir.

---

#### üó≥Ô∏è Adƒ±m 4: **Tahmin ve Oylama**

- Yeni bir veri √∂rneƒüi geldiƒüinde, her aƒüa√ß bir tahminde bulunur.
- T√ºm tahminler birle≈ütirilir:

  - **Sƒ±nƒ±flandƒ±rma i√ßin**: En √ßok tahmin edilen sƒ±nƒ±f se√ßilir.  
    (√∂rneƒüin: [‚ÄòA‚Äô, ‚ÄòA‚Äô, ‚ÄòB‚Äô, ‚ÄòA‚Äô, ‚ÄòB‚Äô] ‚Üí Tahmin = ‚ÄòA‚Äô)
  
  - **Regresyon i√ßin**: Tahminlerin ortalamasƒ± alƒ±nƒ±r.  
    (√∂rneƒüin: [5.2, 4.9, 5.1] ‚Üí Tahmin = 5.07)

---

### üìå Neden Random Forest Daha G√º√ßl√ºd√ºr?

| √ñzellik | A√ßƒ±klama |
|--------|----------|
| ‚úÖ √áe≈üitlilik | Her aƒüa√ß farklƒ± veri ve √∂zelliklerle eƒüitilir ‚Üí farklƒ± √∂ƒürenme |
| ‚úÖ D√º≈ü√ºk varyans | Aƒüa√ßlar birlikte karar verince tek bir aƒüacƒ±n hatasƒ± bastƒ±rƒ±lƒ±r |
| ‚úÖ A≈üƒ±rƒ± √∂ƒürenmeye dayanƒ±klƒ± | Karar aƒüa√ßlarƒ± overfit edebilir, ama RF genel resmi daha iyi yakalar |
| ‚úÖ √ñzellik √∂nemini g√∂sterir | Hangi √∂zelliklerin tahminde ne kadar etkili olduƒüunu g√∂rebiliriz |

---

### üß† Sezgisel Anlatƒ±m √ñrneƒüi: "Sƒ±nƒ±f Temsilcisi Se√ßimi"

Bir sƒ±nƒ±fta temsilci se√ßilecek.  
Her √∂ƒürenci ailesiyle, arkada≈ülarƒ±yla ve √∂ƒüretmeniyle konu≈üarak bir aday belirliyor (farklƒ± bilgiler ‚Üí farklƒ± kararlar).  
Sonra herkes oy kullanƒ±yor ve en √ßok oy alan ki≈üi se√ßiliyor.  
‚Üí Bu ≈üekilde topluluƒüun geneline daha uygun bir karar veriliyor.

## üß† 1. Ensemble Learning (Topluluk √ñƒürenmesi)

> **Tanƒ±m:** Birden fazla makine √∂ƒürenmesi modelini bir araya getirerek, tek bir g√º√ßl√º model olu≈üturma yakla≈üƒ±mƒ±dƒ±r.  
> **Ama√ß:** Bireysel modellerin zayƒ±flƒ±klarƒ±nƒ± dengeleyip daha iyi performans elde etmek.  
> **√ñrnekler:** Bagging, Boosting, Stacking

---

## üß∫ 2. Bagging (Bootstrap Aggregating)

> **Tanƒ±m:** Verinin rastgele alt k√ºmeleriyle (bootstrap √∂rnekleme) farklƒ± modeller eƒüitilir ve tahminler birle≈ütirilir.  
> **Ama√ß:** Varyansƒ± azaltmak, overfitting‚Äôi √∂nlemek.  
> **Klasik √ñrnek:** Random Forest  
> **Tahmin:** Sƒ±nƒ±flandƒ±rma ‚Üí √ßoƒüunluk oyu | Regresyon ‚Üí ortalama

---

## üöÄ 3. Boosting

> **Tanƒ±m:** Zayƒ±f modeller art arda eƒüitilir; her model bir √∂ncekinde yapƒ±lan hatalarƒ± d√ºzeltmeye √ßalƒ±≈üƒ±r.  
> **Ama√ß:** Bias‚Äôƒ± azaltmak, daha g√º√ßl√º bir model olu≈üturmak.  
> **√ñrnek Algoritmalar:** AdaBoost, Gradient Boosting, XGBoost, LightGBM  
> **√ñzellik:** Modeller sƒ±ralƒ± ve birbirine baƒülƒ±dƒ±r (farklƒ± olarak bagging'de modeller paraleldir)

---

## üß± 4. Stacking

> **Tanƒ±m:** Farklƒ± algoritmalar bir arada eƒüitilir, √ßƒ±ktƒ±lar ba≈üka bir modelle birle≈ütirilir (meta-learner).  
> **Ama√ß:** Farklƒ± modellerin g√º√ßl√º yanlarƒ±nƒ± birle≈ütirmek.  
> **Yapƒ±:** Taban modeller + Meta model  
> **√ñrnek:** Lojistik regresyon + karar aƒüacƒ± + kNN ‚Üí √ºstte bir regresyon modeli

---

## üó≥Ô∏è 5. Voting Classifier

> **Tanƒ±m:** Farklƒ± modellerin tahminleri alƒ±nƒ±r, en √ßok √∂nerilen sƒ±nƒ±f se√ßilir.  
> **Y√∂ntemler:**  
> - **Hard Voting:** Sƒ±nƒ±f oylamasƒ±  
> - **Soft Voting:** Sƒ±nƒ±f olasƒ±lƒ±klarƒ±nƒ±n ortalamasƒ±

---

## üß© 6. Feature Bagging

> **Tanƒ±m:** Her aƒüacƒ±n her d√ºƒü√ºm√ºnde t√ºm √∂zellikler yerine rastgele bir alt k√ºme kullanƒ±lmasƒ±.  
> **Random Forest‚Äôƒ±n temel √∂zelliƒüidir.**  
> **Ama√ß:** Aƒüa√ßlarƒ±n farklƒ± kararlar vermesini saƒülamak (√ße≈üitlilik)

---

## üéØ 7. Out-of-Bag Error (OOB Error)

> **Tanƒ±m:** Bootstrap sƒ±rasƒ±nda se√ßilmeyen √∂rnekler (yakla≈üƒ±k %30), modelin doƒüruluƒüunu test etmek i√ßin kullanƒ±lƒ±r.  
> **Avantaj:** Cross-validation yapmadan doƒüruluk tahmini yapƒ±labilir.



---

## üéØ Out-of-Bag Error (OOB Error) Nedir?

### üîπ √ñnce hatƒ±rlayalƒ±m: Bagging ne yapƒ±yordu?

- Random Forest‚Äôta her aƒüa√ß **bootstrap** y√∂ntemiyle eƒüitilir.
- Yani eƒüitim verisinden **geri koymalƒ±** olarak √∂rnekler se√ßilir.
- Bu demektir ki bazƒ± g√∂zlemler **birden fazla kez se√ßilir**, bazƒ±larƒ± ise **hi√ß se√ßilmez**.

üìå Ortalama olarak, her bootstrap √∂rneƒüinde veri setinin **%63'√º** kullanƒ±lƒ±r.  
Geriye kalan **%37‚Äôlik kƒ±sƒ±m** ‚Üí o aƒüaca **‚Äúhi√ß g√∂sterilmemi≈ü‚Äù** olur.

---

### üí° ƒ∞≈üte burada devreye giriyor: OOB (Out-of-Bag)

> **Fikir:**  
> Eƒüer bir g√∂zlem bir aƒüacƒ±n eƒüitiminde *yer almadƒ±ysa*, o aƒüa√ß bu g√∂zlem i√ßin *baƒüƒ±msƒ±z bir test* i≈ülevi g√∂rebilir.

Yani:
- Bir g√∂zlem i√ßin, onu g√∂rmeyen aƒüa√ßlarƒ±n tahminlerine bakƒ±lƒ±r.
- Bu tahminler birle≈ütirilir (oylama/ortalama).
- Sonu√ßla ger√ßek sƒ±nƒ±f kar≈üƒ±la≈ütƒ±rƒ±lƒ±r.
- B√∂ylece **modelin doƒüruluƒüu** hesaplanabilir ‚Üí *OOB Error*.

---

### üìä Avantajƒ± Ne?

‚úÖ **Cross-validation yapmadan doƒüruluk tahmini saƒülar**
* OOB iyi bir ‚Äúerken uyarƒ± sistemi‚Äù gibidir.
* CV ise ‚Äútam saƒülƒ±k taramasƒ±‚Äùdƒ±r.
‚úÖ Ekstra zaman harcamadan, modelin dƒ±≈üsal ba≈üarƒ±mƒ±nƒ± √∂l√ßer  
‚úÖ √ñzellikle b√ºy√ºk veri setlerinde √ßok faydalƒ±dƒ±r

---

### üîç Biraz Kod G√∂zlemi

```python
from sklearn.ensemble import RandomForestClassifier

# OOB tahminlerini aktif edelim
model = RandomForestClassifier(n_estimators=100, oob_score=True, random_state=42)
model.fit(X_train, y_train)

# OOB skoru
print("Out-of-Bag Score:", model.oob_score_)
```

> Bu √ßƒ±ktƒ±, modelin **cross-validation olmadan** elde ettiƒüi test skoruna benzer bir sonu√ß verir.

---

## üß† √ñzetle OOB Error:
- Random Forest‚Äôƒ±n i√ßinde **yerle≈üik bir test y√∂ntemi** gibidir.
- Modelin **daha √∂nce g√∂rmediƒüi** verilerle ne kadar iyi performans g√∂sterdiƒüini tahmin eder.
- √ñzellikle hƒ±zlƒ± doƒürulama ve model kar≈üƒ±la≈ütƒ±rmasƒ± i√ßin √ßok kullanƒ±≈ülƒ±dƒ±r.

---


## ‚öñÔ∏è 8. Bias-Variance Trade-off

> **Tanƒ±m:** Basit modeller y√ºksek bias (yanlƒ±lƒ±k), kompleks modeller y√ºksek varyans hatasƒ± yapar.  
> **Ama√ß:** Bu ikisi arasƒ±nda doƒüru dengeyi bulmak.  
> **Random Forest:** Varyansƒ± azaltƒ±r, bias‚Äôƒ± √ßok artƒ±rmaz ‚Üí dengelidir.

---

## üî• 9. Overfitting & Underfitting

- **Overfitting (A≈üƒ±rƒ± √ñƒürenme):** Model veriye ezber yapar, genelleme yapamaz.  
- **Underfitting (Yetersiz √ñƒürenme):** Model √ßok basittir, veriyi √∂ƒürenemez.  
- **Random Forest genelde overfitting‚Äôi azaltmak i√ßin kullanƒ±lƒ±r.**

---

## üìâ 10. Variance Reduction

> **Tanƒ±m:** Bagging ile aynƒ± veride farklƒ± kararlar veren modellerin ortalamasƒ± alƒ±narak hata dalgalanmasƒ± azaltƒ±lƒ±r.  
> **Random Forest‚Äôƒ±n g√ºc√º burada yatar.**

---

## üé≤ 11. Randomization (Rastgelelik)

> **Tanƒ±m:** Verideki √∂rnekler ve √∂zelliklerin rastgele se√ßilmesi, modelin √ße≈üitliliƒüini artƒ±rƒ±r.  
> **Yarar:** Aƒüa√ßlar aynƒ± veriyle aynƒ± sonu√ßlarƒ± √ºretmez ‚Üí overfitting riski d√º≈üer.

---


````python
# Random Forest ile 24 √ñzellikli Veri Setinde Eƒüitim S√ºreci

from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Veri Seti Olu≈üturma (24 √∂zellikli sentetik sƒ±nƒ±flandƒ±rma verisi)
X, y = make_classification(n_samples=1000, n_features=24, n_informative=10, n_redundant=5,
                           n_clusters_per_class=2, random_state=42)

# Eƒüitim ve test ayrƒ±mƒ±
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Random Forest Modeli
model = RandomForestClassifier(
    n_estimators=100,
    max_features="sqrt",  # her d√ºƒü√ºmde sqrt(24) ‚âà 4.9 ‚Üí yakla≈üƒ±k 5 √∂zellik kullanƒ±lƒ±r
    random_state=42
)
model.fit(X_train, y_train)

# 3. Tahmin ve Performans Deƒüerlendirmesi
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 4. √ñzellik √ñnemleri (Feature Importances)
feature_names = [f"feature_{i}" for i in range(24)]
importances = pd.Series(model.feature_importances_, index=feature_names)

plt.figure(figsize=(10, 6))
sns.barplot(x=importances.sort_values(ascending=False), y=importances.sort_values(ascending=False).index)
plt.title("√ñzellik √ñnemleri (Feature Importances)")
plt.xlabel("√ñnem Skoru")
plt.ylabel("√ñzellikler")
plt.tight_layout()
plt.show()

# 5. Tek bir aƒüacƒ±n g√∂rselle≈ütirilmesi (isteƒüe baƒülƒ±)
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 10))
plot_tree(model.estimators_[0], max_depth=3, feature_names=feature_names, filled=True, fontsize=10)
plt.title("Random Forest ƒ∞√ßindeki Bir Aƒüa√ß (ƒ∞lk Aƒüa√ß)")
plt.show()

````

````python
# Random Forest ile Iris Veri Setinde Eƒüitim, Aƒüa√ß G√∂rselle≈ütirme ve OOB vs CV Kar≈üƒ±la≈ütƒ±rmasƒ±

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, accuracy_score
from sklearn.tree import plot_tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

````python

# 1. Veri Seti Y√ºkleme
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
````

````python
# Eƒüitim ve test ayrƒ±mƒ±
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
# 2. Random Forest Modeli (OOB etkin)
model_oob = RandomForestClassifier(
    n_estimators=100,
    max_features="sqrt",
    oob_score=True,
    random_state=42
)
model_oob.fit(X_train, y_train)
````

````python
# 3. OOB Skoru ve Test Ba≈üarƒ±mƒ±
print("OOB Score:", model_oob.oob_score_)
print("Test Accuracy:", accuracy_score(y_test, model_oob.predict(X_test)))
````

````python
# 4. Cross-Validation ile Performans √ñl√ß√ºm√º
cv_scores = cross_val_score(model_oob, X, y, cv=5)
print("CV Scores:", cv_scores)
print("CV Mean Accuracy:", np.mean(cv_scores))
````

````python
# 5. √ñzellik √ñnemleri (Feature Importances)
importances = pd.Series(model_oob.feature_importances_, index=feature_names)
plt.figure(figsize=(8, 5))
sns.barplot(x=importances.sort_values(ascending=False), y=importances.sort_values(ascending=False).index)
plt.title("√ñzellik √ñnemleri (Feature Importances)")
plt.xlabel("√ñnem Skoru")
plt.ylabel("√ñzellikler")
plt.tight_layout()
plt.show()
````

````python
# 6. Birden Fazla Aƒüacƒ±n G√∂rselle≈ütirilmesi
for i in range(3):
    plt.figure(figsize=(16, 8))
    plot_tree(
        model_oob.estimators_[i],
        max_depth=3,
        feature_names=feature_names,
        class_names=iris.target_names,
        filled=True,
        fontsize=10
    )
    plt.title(f"Random Forest - Aƒüa√ß {i+1}")
    plt.show()
````

````python
# 7. K√∂k √ñzellikleri G√∂zlemi (ƒ∞lk 5 Aƒüa√ß)
print("--- K√∂k D√ºƒü√ºm √ñzellikleri ---")
for i, tree in enumerate(model_oob.estimators_[:]):
    root_feature_index = tree.tree_.feature[0]
    print(f"Aƒüa√ß {i+1} k√∂k √∂zelliƒüi: {feature_names[root_feature_index]}")
````

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Veri Setini ƒ∞ndir
path = "/content/drive/MyDrive/KAIRU/Kurumsal/Acun Medya EgÃÜitimi/datasets/student-por.csv"
df = pd.read_csv(path)
df.head()
````

````python
# 2. Kategorik Deƒüi≈ükenleri Sayƒ±sal Deƒüerlere √áevir
df_encoded = df.copy()
label_encoders = {}
for col in df_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le

#df_ = df_encoded[['G1','G2','G3']]
#X = df_.drop(columns=['G3'])  # Dilersen G1, G2'yi ekleyebilirsin
#y = df_['G3']


# 3. Hedef deƒüi≈üken: G3 (final notu), G1 ve G2 √ßƒ±karƒ±labilir (√∂n bilgi i√ßerdiƒüi i√ßin)

X = df_encoded.drop(columns=['G3'])  # Dilersen G1, G2'yi ekleyebilirsin
y = df_encoded['G3']

# 4. Eƒüitim/Test Ayrƒ±mƒ±
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Modeli Kur ve Eƒüit
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 6. Tahmin ve Deƒüerlendirme
y_pred = model.predict(X_test)
print("R2 Score:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))

# 7. Ger√ßek vs Tahmin Grafiƒüi
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred)
plt.plot([0, 20], [0, 20], '--r')
plt.xlabel("Ger√ßek G3")
plt.ylabel("Tahmin Edilen G3")
plt.title("Random Forest ile G3 Tahmini")
plt.tight_layout()
plt.show()

# 8. √ñzellik √ñnemleri
importances = pd.Series(model.feature_importances_, index=X.columns)
importances.sort_values(ascending=False).head(10).plot(kind='barh')
plt.title("En √ñnemli 10 √ñzellik")
plt.xlabel("√ñnemi")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

Tabii ki, **Parkinsons Telemonitoring** veri setindeki deƒüi≈ükenlerin anlamlarƒ±nƒ± a≈üaƒüƒ±da bulabilirsiniz:

- **subject#**: Her bir hastaya atanmƒ±≈ü benzersiz kimlik numarasƒ±.ÓàÜ

- **age**: Hastanƒ±n ya≈üƒ± (yƒ±l cinsinden).ÓàÜ

- **sex**: Hastanƒ±n cinsiyeti; 0 = erkek, 1 = kadƒ±n.ÓàÜ

- **test_time**: ƒ∞lk ziyaret tarihinden itibaren ge√ßen s√ºre (g√ºn cinsinden).ÓàÜ

- **motor_UPDRS**: Motor Unified Parkinson's Disease Rating Scale (UPDRS) skoru; hastanƒ±n motor fonksiyonlarƒ±nƒ± deƒüerlendiren bir √∂l√ßek.ÓàÜ

- **total_UPDRS**: Toplam UPDRS skoru; hem motor hem de motor dƒ±≈üƒ± semptomlarƒ± kapsayan genel bir deƒüerlendirme.ÓàÜ

- **Jitter(%)**: Ses frekansƒ±ndaki kƒ±sa s√ºreli rastgele dalgalanmalarƒ±n y√ºzdesi; ses titremesinin bir √∂l√ß√ºt√º.ÓàÜ

- **Jitter(Abs)**: Mutlak jitter deƒüeri; ses frekansƒ±ndaki kƒ±sa s√ºreli rastgele dalgalanmalarƒ±n mutlak deƒüeri.ÓàÜ

- **Jitter:RAP**: RAP (Relative Average Perturbation); belirli bir s√ºre boyunca frekans pert√ºrbasyonlarƒ±nƒ±n ortalamasƒ±.ÓàÜ

- **Jitter:PPQ5**: PPQ5 (Pitch Perturbation Quotient); 5 periyotluk bir pencere boyunca frekans pert√ºrbasyonlarƒ±nƒ±n ortalamasƒ±.ÓàÜ

- **Jitter:DDP**: DDP (Divergence of Distinctive Points); jitter RAP deƒüerinin √º√ß katƒ± olarak hesaplanƒ±r.ÓàÜ

- **Shimmer**: Ses genliƒüindeki kƒ±sa s√ºreli rastgele dalgalanmalarƒ±n y√ºzdesi; sesin ≈üiddetindeki titremelerin bir √∂l√ß√ºt√º.ÓàÜ

- **Shimmer(dB)**: Desibel cinsinden shimmer deƒüeri; ses genliƒüindeki dalgalanmalarƒ±n desibel olarak ifadesi.ÓàÜ

- **Shimmer:APQ3**: APQ3 (Amplitude Perturbation Quotient); 3 periyotluk bir pencere boyunca genlik pert√ºrbasyonlarƒ±nƒ±n ortalamasƒ±.ÓàÜ

- **Shimmer:APQ5**: APQ5; 5 periyotluk bir pencere boyunca genlik pert√ºrbasyonlarƒ±nƒ±n ortalamasƒ±.ÓàÜ

- **Shimmer:APQ11**: APQ11; 11 periyotluk bir pencere boyunca genlik pert√ºrbasyonlarƒ±nƒ±n ortalamasƒ±.ÓàÜ

- **Shimmer:DDA**: DDA (Divergence of Distinctive Amplitudes); shimmer APQ3 deƒüerinin √º√ß katƒ± olarak hesaplanƒ±r.ÓàÜ

- **NHR**: Noise-to-Harmonics Ratio; ses sinyalindeki g√ºr√ºlt√º enerjisinin harmonik bile≈üenlere oranƒ±.ÓàÜ

- **HNR**: Harmonics-to-Noise Ratio; harmonik bile≈üenlerin g√ºr√ºlt√º enerjisine oranƒ±.ÓàÜ

- **RPDE**: Recurrence Period Density Entropy; ses sinyalinin tekrarlayan periyotlarƒ±nƒ±n yoƒüunluk entropisi, sesin karma≈üƒ±klƒ±ƒüƒ±nƒ± √∂l√ßer.ÓàÜ

- **DFA**: Detrended Fluctuation Analysis; ses sinyalinin fraktal √∂l√ßeklenme √∂zelliklerini analiz eder.ÓàÜ

- **PPE**: Pitch Period Entropy; ses periyotlarƒ±nƒ±n entropisi, ses tonunun deƒüi≈ükenliƒüini √∂l√ßer.ÓàÜ

Bu deƒüi≈ükenler, Parkinson hastalƒ±ƒüƒ±nƒ±n ilerleyi≈üini ve semptomlarƒ±n ≈üiddetini deƒüerlendirmek i√ßin kullanƒ±lƒ±r. √ñzellikle sesle ilgili √∂l√ß√ºmler, hastalƒ±ƒüƒ±n motor semptomlarƒ±nƒ±n uzaktan izlenmesi a√ßƒ±sƒ±ndan √∂nemlidir.

````python

from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler



# Gerekli s√ºtunlarƒ± se√ß (hedef ve kimlik bilgileri hari√ß)
X = df.drop(columns=["subject#"])

# VIF hesaplamasƒ± i√ßin √∂nce √∂l√ßeklendirme yap
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# VIF hesapla
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]
# Bilimsel g√∂sterimi kapat (float formatƒ±nƒ± belirle)
pd.options.display.float_format = '{:,.2f}'.format

# Sonu√ßlarƒ± sƒ±rala ve g√∂ster
vif_data_sorted = vif_data.sort_values(by="VIF", ascending=False)
print(vif_data_sorted)
````

````python
#3. Hedef ve √ñzellikleri VIF'e g√∂re ayarla (VIF hesapladƒ±ktan sonra √ßalƒ±≈ütƒ±r.)
# 'motor_UPDRS' tahmin edilecek deƒüi≈üken, ki≈üi bilgilerini hari√ß tutuyoruz
X = df.drop(columns=['subject#','motor_UPDRS', 'total_UPDRS', 'test_time',
    "Jitter:RAP",
    "Jitter:PPQ5",
    "Jitter:DDP",
    "Shimmer(dB)",
    "Shimmer:APQ3",
    "Shimmer:APQ5",
    "Shimmer:APQ11",
    "Shimmer:DDA",
    "total_UPDRS",
    "age_group",
    ])

y = df['motor_UPDRS']
````

````python
# 4. Eƒüitim ve Test Verisine Ayƒ±r
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Modeli Kur ve Eƒüit
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 6. Tahmin ve Deƒüerlendirme
y_pred = model.predict(X_test)
print("R2 Score:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))


````

---

## üß† SHAP Nedir?

SHAP, bir modelin tahminini **√∂zellik bazƒ±nda a√ßƒ±klayan** bir y√∂ntemdir.

> ‚ÄúModel bu tahmini neden yaptƒ±?‚Äù  
> ‚ÄúHangi √∂zellik bu tahmini artƒ±rdƒ± veya azalttƒ±?‚Äù  
> ‚ÄúBu hasta neden y√ºksek riskli olarak sƒ±nƒ±flandƒ±rƒ±ldƒ±?‚Äù  

Bu t√ºr sorulara **matematiksel olarak adil ve g√ºvenilir bir ≈üekilde** yanƒ±t verir.

---

## üìä SHAP Ne Yapar?

Modelin her tahmini i√ßin:

- **Beklenen tahmin (ortalama)** ‚Üí `explainer.expected_value`
- **Tahmin edilen deƒüer** ‚Üí `model.predict(...)`
- **Her √∂zelliƒüin tahmine katkƒ±sƒ±** ‚Üí `SHAP deƒüeri`

üîÅ Toplamda:
```python
model output ‚âà base_value + ‚àë SHAP contributions
```
