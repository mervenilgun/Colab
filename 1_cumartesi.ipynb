

## ✅ **Yapay Zekâya Giriş: Tanım, Tarihçe ve Alt Dalları**

---

### 🧠 **1. Yapay Zekâ (Artificial Intelligence - AI) Nedir?**

**Tanım:**
Yapay Zekâ, insan zekâsının bilgisayar sistemleri tarafından taklit edilmesidir. Öğrenme, akıl yürütme, problem çözme, dil anlama, algılama gibi bilişsel işlevlerin yazılım ve algoritmalarla yapılmasını sağlar.

**Basitçe:**

> "Yapay zekâ, bilgisayarların ‘zeki’ gibi davranmasıdır."

---

### 📜 **2. Kısaca Yapay Zekâ Tarihçesi**

* **1950 – Alan Turing:** “Makineler düşünebilir mi?” sorusu ve **Turing Testi**
* **1956 – Dartmouth Konferansı:** “Artificial Intelligence” terimi ilk kez kullanıldı
* **1980'ler – Bilgi Tabanlı Sistemler (Expert Systems)**
* **2000'ler – Veri patlaması, GPU’lar ve internetle birlikte büyük ilerlemeler**
* **2020'ler – LLM’ler (Büyük Dil Modelleri), ChatGPT, DALL·E vb.**

---

### 🧭 **3. Yapay Zekânın Üç Ana Kategorisi**

| Tür                | Açıklama                          | Örnek                  |
| ------------------ | --------------------------------- | ---------------------- |
| **Dar (Zayıf) AI** | Tek bir görevi yapar              | Siri, spam filtresi    |
| **Genel AI**       | Her alanda insan gibi düşünebilir | Henüz yok              |
| **Süper AI**       | İnsan zekâsını aşar               | Kuramsal (bilim kurgu) |

---

### 🌿 **4. Yapay Zekânın Alt Dalları**

Yapay zekâ şeması:

```
Yapay Zekâ (AI)
 ├── Makine Öğrenmesi (ML)
 │    ├── Denetimli Öğrenme
 │    ├── Denetimsiz Öğrenme
 │    └── Pekiştirmeli Öğrenme
 ├── Derin Öğrenme (DL)
 ├── Doğal Dil İşleme (NLP)
 ├── Bilgisayarla Görü (CV)
 └── Uzman Sistemler (Expert Systems)
```

---

### 🔍 **5. Makine Öğrenmesi (Machine Learning - ML)**

**Tanım:**
Verilerden otomatik olarak öğrenen ve tahmin yapan algoritmalar bütünüdür.

> "Kodlama değil, verilerle öğrenme."

#### 📘 **3 Alt Başlık:**

| Tür                      | Açıklama                   | Örnek                 |
| ------------------------ | -------------------------- | --------------------- |
| **Denetimli Öğrenme**    | Girdi + Çıktı verisi var   | E-posta spam tespiti  |
| **Denetimsiz Öğrenme**   | Sadece girdi verisi var    | Müşteri segmentasyonu |
| **Pekiştirmeli Öğrenme** | Deneme-yanılma ile öğrenme | Oyun oynayan robot    |

---

### 🌊 **6. Derin Öğrenme (Deep Learning - DL)**

**Tanım:**
İnsan beynini taklit eden **yapay sinir ağlarıyla** öğrenme yöntemidir. Özellikle görüntü, ses ve dil verilerinde çok başarılıdır.

#### 📌 Örnekler:

* Görüntüden kedi/köpek ayırt etme
* Sesle asistan komutu anlama
* Otomatik çeviri (Google Translate)

**Görsel:** 3 katmanlı yapay sinir ağı (input-hidden-output)

---

### 🗣️ **7. Doğal Dil İşleme (Natural Language Processing - NLP)**

**Tanım:**
Bilgisayarların insan dilini (Türkçe, İngilizce vb.) anlama ve üretme becerisidir.

#### 📌 Örnekler:

* ChatGPT, Bard, Claude
* E-posta sınıflandırma
* Duygu analizi (mutlu/mutsuz yorumlar)

---

### 👁️ **8. Bilgisayarla Görü (Computer Vision - CV)**

**Tanım:**
Görüntü veya video verilerini analiz etme ve anlama alanıdır.

#### 📌 Örnekler:

* Yüz tanıma sistemleri
* Trafik levhalarını tanıyan araçlar
* Röntgen analizleri

---

### 👩‍⚕️ **9. Uzman Sistemler (Expert Systems)**

**Tanım:**
Belirli bir uzmanlık alanındaki bilgileri kurallarla kodlayan yapay zekâ sistemleridir. 1980’lerde yaygındı.

#### 📌 Örnek:

* Tıbbi teşhis sistemleri
* Satın alma öneri motorları

---

### 🧭 **10. Özetle: Yapay Zekâ Ne Değildir?**

* Yapay zekâ **her şeyi bilen** değildir
* Yalnızca **veriler kadar zekidir**
* "Öğrenen sistem" değildir her zaman – bazıları kurallarla çalışır


## 🧾 Python Veri Tipleri – Temel Türler ve Örnekler

---

### 🔢 1. **Tam Sayı (Integer - `int`)**

**Açıklama:** Ondalık kısmı olmayan sayılar.
**Örnek:**

```python
sayi = 42
print(type(sayi))  # <class 'int'>
```

🧠 Kullanım Alanı: Yaş, miktar, sıra numarası, sayaçlar...

---

### 🌊 2. **Ondalıklı Sayı (Float - `float`)**

**Açıklama:** Ondalıklı yani küsuratlı sayılar.
**Örnek:**

```python
pi = 3.14
print(type(pi))  # <class 'float'>
```

🧠 Kullanım Alanı: Ölçümler, fiyatlar, yüzdeler...

---

### 📝 3. **Metin (String - `str`)**

**Açıklama:** Çift veya tek tırnak arasına yazılan karakterler dizisi.
**Örnek:**

```python
isim = "Ayşe"
print(type(isim))  # <class 'str'>
```

🧠 Kullanım Alanı: İsimler, açıklamalar, adresler, yazılar...

---

### ✅ 4. **Mantıksal (Boolean - `bool`)**

**Açıklama:** Sadece iki değer alır: `True` veya `False`
**Örnek:**

```python
aktif_mi = True
print(type(aktif_mi))  # <class 'bool'>
```

🧠 Kullanım Alanı: Koşullar, kontrol yapıları, seçimler...

---

### 📦 5. **Liste (List - `list`)**

**Açıklama:** Birden fazla değeri sırayla tutan yapılardır. Değiştirilebilir (mutable).
**Örnek:**

```python
renkler = ["kırmızı", "mavi", "yeşil"]
print(type(renkler))  # <class 'list'>
print(renkler[0])     # kırmızı
renkler.append("sarı")  # yeni eleman ekleme
```

🧠 Kullanım Alanı: Öğrenci listesi, alışveriş sepeti, notlar...

---

### 📎 6. **Demet (Tuple - `tuple`)**

**Açıklama:** Listelere benzer ama değiştirilemez (immutable).
**Örnek:**

```python
koordinat = (10, 20)
print(type(koordinat))  # <class 'tuple'>
print(koordinat[1])     # 20
```

🧠 Kullanım Alanı: Sabit veriler, koordinatlar, ayarlar...

---

### 🗃️ 7. **Sözlük (Dictionary - `dict`)**

**Açıklama:** Anahtar-değer çiftleriyle veri saklayan yapıdır.
**Örnek:**

```python
ogrenci = {
    "ad": "Zeynep",
    "yas": 22,
    "aktif": True
}
print(type(ogrenci))       # <class 'dict'>
print(ogrenci["yas"])      # 22
```

🧠 Kullanım Alanı: Profil bilgileri, yapılandırmalar...

---

### 📦 8. **Küme (Set - `set`)**

**Açıklama:** Benzersiz (tekrarsız) değerlerden oluşur, sırasızdır.
**Örnek:**

```python
sayilar = {1, 2, 3, 3, 2}
print(type(sayilar))  # <class 'set'>
print(sayilar)        # {1, 2, 3}
```

🧠 Kullanım Alanı: Yinelenmeyen öğeleri saklamak, kümelerle işlem yapmak...

## 🔁 **while Döngüsü**

```python
# 0'dan 4'e kadar sayıları yazdırma
sayi = 0

while sayi < 5:
    print(sayi)
    sayi += 1
```

```python
# Kullanıcı doğru şifre girene kadar döngü
sifre = ""

while sifre != "1234":
    sifre = input("Şifreyi girin: ")
print("Doğru şifre!")
```




````python
# Örnek 1: Pozitif, negatif veya sıfır kontrolü
sayi = -5
if sayi > 0:
    print("Pozitif")
elif sayi == 0:
    print("Sıfır")
else:
    print("Negatif")
````

````python
# Örnek 2: Yaş kontrolü
yas = 18
if yas < 18:
    print("Reşit değilsiniz.")
else:
    print("Reşitsiniz.")
````

````python
# Örnek 3: Liste elemanlarını yazdırma
meyveler = ["elma", "muz", "çilek"]
for meyve in meyveler:
    print(meyve)
````

````python
# Örnek 4: 1'den 5'e kadar olan sayılar
for i in range(1, 6):
    print(i)
````

````python
# Örnek 5: 1'den 10'a kadar çift sayılar
for i in range(1, 11):
    if i % 2 == 0:
        print(i, "çift sayıdır.")
````

````python
# Örnek 6: Sayaç örneği
sayi = 0
while sayi < 5:
    print(sayi)
    sayi += 1
````

````python
# Örnek 7: Şifre kontrolü (simülasyonlu)
sifre = "0000"
deneme = "1234"
while sifre != deneme:
    print("Yanlış şifre:", sifre)
    sifre = "1234"  # (normalde input() kullanılır)
print("Doğru şifre!")
````

````python
# Örnek 7: Şifre kontrolü (input ile, 3 deneme hakkı)
dogru_sifre = "1234"
hak = 3

while hak > 0:
    sifre = input("Şifreyi girin: ")

    if sifre == dogru_sifre:
        print("Doğru şifre! Giriş başarılı.")
        break
    else:
        hak -= 1
        print("Yanlış şifre. Kalan deneme hakkı:", hak)

if hak == 0:
    print("Hesabınız kilitlendi. Lütfen daha sonra tekrar deneyin.")
````

**try-except yapısı**

````python
try:
    sayi = int(input("Bir sayı girin: "))

    if sayi > 0:
        print("Pozitif bir sayı")
    elif sayi < 0:
        print("Negatif bir sayı")
    else:
        print("Sayı sıfır")

except ValueError:
    print("Lütfen sadece rakam girin! Harf veya özel karakter kabul edilmez.")
````

5.Uygulama: Girilen sayıya kadar sadece çift sayıları yazdırsın

6.Uygulama: Girilen sayıya kadar tek-çift sayıları yazdıran

````python
n = int(input("Kaça kadar yazılsın? "))

for i in range(1, n + 1):
    if i % 2 == 0:
        print(i, "çifttir")
    else:
        print(i, "tektir")
````
